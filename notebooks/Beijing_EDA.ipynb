```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# TODO: Use python PathLib to use 
DATA_PATH = '../data/'
```

```python
beijing_aq_data = pd.read_feather(DATA_PATH + 'interim/beijing_aq_union.feather')
```

```python
beijing_monthly_pollutants = beijing_aq_data.set_index('utc_datetime').tz_localize('Asia/Shanghai')[['PM2.5', 'PM10', 'O3']].resample('M').mean().to_period('M')
```

## Beijing monthly mean pollutant concentration


https://www.ncbi.nlm.nih.gov/pmc/articles/PMC6147404/

```python
fig, ax = plt.subplots()
beijing_monthly_pollutants.plot(figsize=(10, 5), ax=ax)
ax.set_title('Beijing Monthly Mean Pollutant COncentration')
ax.set_xlabel('Month')
ax.set_ylabel('micrograms/m3')

fig.savefig('beijing_monthly_pollutants.png')
```
